Univariate Analysis

```python
#Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Monthly Income
plt.figure(figsize=(10, 6))
sns.histplot(df['Monthly Income'], bins=30, kde=True)
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.show()
```

```python
#Count of Attrition
plt.figure(figsize=(10, 6))
sns.countplot(x='Attrition', data=df)
plt.title('Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

Bivariate Analysis

```python
#Attrition by Department
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
#Attrition by Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Monthly Income vs. Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Monthly Income', hue='Attrition', data=df)
plt.title('Monthly Income vs. Age')
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.show()
```

Multivariate Analysis

```python
#Pairplot of Selected Features
selected_features = ['Age', 'Monthly Income', 'Total Working Years', 'Attrition']
sns.pairplot(df[selected_features], hue='Attrition')
plt.title('Pairplot of Selected Features')
plt.show()
```

Feature Engineering

```python
#Creating Age Bands
age_bins = [18, 25, 35, 45, 55, 65]
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-65']
df['Age Band'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

plt.figure(figsize=(10, 6))
sns.countplot(x='Age Band', hue='Attrition', data=df)
plt.title('Attrition by Age Band')
plt.xlabel('Age Band')
plt.ylabel('Count')
plt.show()
```

```python
#Creating a Feature for Years Since Last Promotion
df['Years Since Last Promotion'] = df['Years At Company'] - df['Years In Current Role']

plt.figure(figsize=(10, 6))
sns.histplot(df['Years Since Last Promotion'], bins=20, kde=True)
plt.title('Years Since Last Promotion Distribution')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Frequency')
plt.show()
```
